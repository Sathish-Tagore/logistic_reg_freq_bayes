```python
feature_names1 = [int(beta_name.split('.')[1]) - 1 for beta_name in features_set1]
feature_names1 = df_X_train.columns[feature_names1]
feature_names2 = [int(beta_name.split('.')[1]) - 1 for beta_name in features_set2]
feature_names2 = df_X_train.columns[feature_names2]
```

```python
log_summary1 = logistic.loc[:,features_set1].T
log_summary1 = log_summary1.set_index(feature_names1)

log_feat_dict1 = {}
for i, row in log_summary1.iterrows():
    log_feat_dict1[i] = list(np.array(log_summary1.loc[i,:]))


lasso_summary1 = lasso.loc[:,features_set1].T
lasso_summary1 = lasso_summary1.set_index(feature_names1)

lasso_feat_dict1 = {}
for i, row in lasso_summary1.iterrows():
    lasso_feat_dict1[i] = list(np.array(lasso_summary1.loc[i,:]))

elasticnet_summary1 = elasticnet.loc[:,features_set1].T
elasticnet_summary1 = elasticnet_summary1.set_index(feature_names1)

elasticnet_feat_dict1 = {}
for i, row in elasticnet_summary1.iterrows():
    elasticnet_feat_dict1[i] = list(np.array(elasticnet_summary1.loc[i,:]))


horseshoe_summary1 = horseshoe.loc[:,features_set1].T
horseshoe_summary1 = horseshoe_summary1.set_index(feature_names1)

hs_feat_dict1 = {}
for i, row in horseshoe_summary1.iterrows():
    hs_feat_dict1[i] = list(np.array(horseshoe_summary1.loc[i,:]))
```

```python
log_summary2 = logistic.loc[:,features_set2].T
log_summary2 = log_summary2.set_index(feature_names2)

log_feat_dict2 = {}
for i, row in log_summary2.iterrows():
    log_feat_dict2[i] = list(np.array(log_summary2.loc[i,:]))


lasso_summary2 = lasso.loc[:,features_set2].T
lasso_summary2 = lasso_summary2.set_index(feature_names2)

lasso_feat_dict2 = {}
for i, row in lasso_summary2.iterrows():
    lasso_feat_dict2[i] = list(np.array(lasso_summary2.loc[i,:]))

elasticnet_summary2 = elasticnet.loc[:,features_set2].T
elasticnet_summary2 = elasticnet_summary2.set_index(feature_names2)

elasticnet_feat_dict2 = {}
for i, row in elasticnet_summary2.iterrows():
    elasticnet_feat_dict2[i] = list(np.array(elasticnet_summary2.loc[i,:]))


horseshoe_summary2 = horseshoe.loc[:,features_set2].T
horseshoe_summary2 = horseshoe_summary2.set_index(feature_names2)

hs_feat_dict2 = {}
for i, row in horseshoe_summary2.iterrows():
    hs_feat_dict2[i] = list(np.array(horseshoe_summary2.loc[i,:]))
```
